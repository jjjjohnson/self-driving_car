```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_radar = pd.read_csv('NIS_radar.csv',delimiter='\t')
df_radar['95%'] = 7.815
```

```python
plt.plot(df_radar['Count'], df_radar['NIS'])
plt.plot(df_radar['Count'], df_radar['95%'])
plt.ylim(0,20)
```
